```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.arima_model import ARMA
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose


```

```python
AMZN = "/Users/davidhoffman/desktop/stocks_download/AMZN_2006-01-01_to_2018-01-01.csv"
df = pd.read_csv(AMZN, index_col="Date", parse_dates=True)
```

```python
#Only keep closing price of the day
df.drop(['Open','High','Low'], axis = 1, inplace = True)
df.drop(['Name'], axis = 1, inplace = True)
df.drop(['Volume'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# In time series data, autocorrelation is a measure of how correlated a variable is with itself.
#Specifically, autocorrelation measures how closely earlier values relate to values that occur later in time.
#To compute autocorrelation, we fix a lag, k, which is the delta between the given point and 
#the prior point used to compute the correlation.

#With a k value of 1, we'd compute how correlated a value is with the value that's immediately prior. 
#With a k value of 10, we'd compute how correlated a variable is with the value that's 10 time points prior.
```

```python
autocorr_quarter_lag1 = df.Close.resample('Q').mean().autocorr(lag=1)
autocorr_year_lag1 = df.Close.resample('A').mean().autocorr(lag=1)

print(autocorr_quarter_lag1)
print(autocorr_year_lag1)
```

```python
#High correlation
```

```python
#resampling and the letter chosen explained
Alias	Description
B	Business day
D	Calendar day
W	Weekly
M	Month end
Q	Quarter end
A	Year end
BA	Business year end
AS	Year start
H	Hourly frequency
T, min	Minutely frequency
S	Secondly frequency
L, ms	Millisecond frequency
U, us	Microsecond frequency
N, ns	Nanosecond frequency
```

```python
fig, ax = plt.subplots(figsize=(12,5))
plot_acf(df.Close, lags=30, ax=ax)
plt.show();

#What does this mean exactly?
```

```python
#Partial Autocorrelation

#Another important chart for diagnosing your time series is the partial autocorrelation chart (PACF). 
#This is similar to autocorrelation, but, instead of being just the correlation at increasing lags, 
#it is the correlation at a given lag, controlling for the effect of previous lags.
```

```python
fig, ax = plt.subplots(figsize=(12,5))
plot_pacf(df.Close, lags=30, ax=ax)
plt.show();
```

```python
# time series decomposition
#Questions:

#What is the long-term behavior of my series?
#How does my time series fluctuate?

#We can create a time series comprised of a linearly increasing trend from 1 to 99 
#and some random noise and decompose it as an additive model.

result = seasonal_decompose(df.Close, model='additive', freq=1)

result.plot()
```

```python
#exploring what the graphs look like under a different model type
#We can contrive a quadratic time series as a square of the time step from 1 to 99, 
#and then decompose it assuming a multiplicative model.

result = seasonal_decompose(df.Close, model='multiplicative', freq=1)

result.plot()
```
