```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.arima_model import ARMA
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import r2_score

```

```python
GS = "/Users/davidhoffman/desktop/stocks_download/GS_2006-01-01_to_2018-01-01.csv"
df = pd.read_csv(GS, index_col="Date", parse_dates=True)
```

```python
#Only keep closing price of the day
df.drop(['Open','High','Low'], axis = 1, inplace = True)
df.drop(['Name'], axis = 1, inplace = True)
df.drop(['Volume'], axis = 1, inplace = True)
```

```python
df['close_diff_1']=df.Close.diff()
```

- Goal: analyze predictions of: 
    -AR vs ARIMA model
    -Using close vs close_diff
    -Out of sample vs in sample predictions

```python
#AR model
ar1 = ARMA(df.Close.dropna().values, (1, 0)).fit()
ar1_diff = ARMA(df.close_diff_1.dropna().values, (1, 0)).fit()

#ARIMA model

ar1ma1 = ARIMA(df.Close.dropna().values, (1,1,1)).fit()
ar1ma1_diff = ARIMA(df.close_diff_1.dropna().values, (1,1,1)).fit()
```

```python
#Predictions in sample

predictions_ar = ar1.predict()

predictions_ar_diff = ar1_diff.predict()
full_pred_ar_diff = df.Close.values[0]+np.cumsum(ar1_diff.fittedvalues)

predictions_ar1ma1 = ar1ma1.predict()
full_pred_ar_ar1ma1 = df.Close.values[0]+np.cumsum(ar1ma1.fittedvalues)

predictions_ar1ma1_diff = ar1ma1_diff.predict()
full_pred_ar1ma1_diff = df.Close.values[0]+np.cumsum(ar1ma1_diff.fittedvalues)
```

```python
#Actual close price vs predictions_ar, in sample

fig, ax = plt.subplots(figsize=(12,5))
date_ticks=df.index
ax.plot(date_ticks, predictions_ar, lw=2, color='grey', ls='dashed', label = 'Predictions')
ax.plot(date_ticks, df.Close, lw=2, color='darkred', label = 'Actual')
plt.xlabel("Date")
plt.ylabel("Closing price")
plt.title("Close vs predicted close")
plt.legend()
plt.show()
```

```python
#Actual close price vs full_pred_ar_diff in sample

fig, ax = plt.subplots(figsize=(12,5))
date_ticks=df.index
ax.plot(date_ticks[1:], full_pred_ar_diff, lw=2, color='grey', ls='dashed', label = 'Predictions')
ax.plot(date_ticks, df.Close, lw=2, color='darkred', label = 'Actual')
plt.xlabel("Date")
plt.ylabel("Closing price")
plt.title("Close vs predicted close")
plt.legend()
plt.show();
```

```python
#Actual close price vs full_pred_ar_ar1ma1 in sample

fig, ax = plt.subplots(figsize=(12,5))
date_ticks=df.index
ax.plot(date_ticks[1:], full_pred_ar_ar1ma1, lw=2, color='grey', ls='dashed', label = 'Predicted')
ax.plot(date_ticks[1:], df.Close[1:], lw=2, color='darkred', label = 'Actual')
plt.xlabel("Date")
plt.ylabel("Closing price")
plt.title("Close vs predicted close")
plt.legend()
plt.show();
```

```python
#Actual close price vs full_pred_ar1ma1_diff in sample

fig, ax = plt.subplots(figsize=(12,5))
date_ticks=df.index
ax.plot(date_ticks[2:], full_pred_ar1ma1_diff, lw=2, color='grey', ls='dashed', label = 'Predicted')
ax.plot(date_ticks[1:], df.Close[1:], lw=2, color='darkred', label = 'Actual')
plt.xlabel("Date")
plt.ylabel("Closing price")
plt.title("Close vs predicted close")
plt.legend()
plt.show()
```
