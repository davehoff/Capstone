```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.arima_model import ARMA
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import r2_score

```

```python
VZ = "/Users/davidhoffman/desktop/stocks_download/VZ_2006-01-01_to_2018-01-01.csv"
df = pd.read_csv(VZ, index_col="Date", parse_dates=True)
```

```python
#Only keep closing price of the day
df.drop(['Open','High','Low'], axis = 1, inplace = True)
df.drop(['Name'], axis = 1, inplace = True)
df.drop(['Volume'], axis = 1, inplace = True)
```

```python
#Check for stationarity
#Perform Dicky Fuller test to test for stationarity
# Function taken from class lecture


def test_stationarity(timeseries):

    # Determining rolling statistics:
    rolmean = timeseries.rolling(window=12, center=False).mean()
    rolstd = timeseries.rolling(window=12, center=False).std()

    # Plotting rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    # Performing a Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries)
    dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
    print(dfoutput )

test_stationarity(df.Close)
```

```python
#Make it stationary by diff
df['close_diff_1']=df.Close.diff()
test_stationarity(df.close_diff_1.dropna())
```

```python
#AR Model
ar1 = ARMA(df.Close.dropna().values, (1, 0)).fit()
ar_preds = ar1.predict()

date_ticks = df.index

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(date_ticks, ar_preds, lw=2, color='grey', ls='dashed')
ax.plot(date_ticks, df.Close, lw=2, color='darkred')
plt.xlabel("Date")
plt.ylabel("Closing price")
plt.title("Close vs predicted close")
plt.legend()
plt.show()
```
